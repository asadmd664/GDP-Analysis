###  Lets calculate mean % growth over the given years

```python
gdp_states['Mean % growth (1314-1516)']=gdp_states['(% Growth over previous year)'].mean(axis=1)
```

```python
gdp_states
```

####  Our target year is 2015-16

```python
target_year = gdp_states.loc[:,[('(% Growth over previous year)','2015-16'),
                                ('GSDP - CURRENT PRICES (` in Crore)','2015-16'),
                                ('Mean % growth (1314-1516)','')]]
```

```python
target_year
```

```python
target_year.columns=['Growth % over previous year','GSDP CURRENT PRICES in Crore', 'Mean % growth (1314-1516)']
```

```python
target_year
```

###  Lets assign acc to states, union territories and overall gdp

```python
gdp_st = target_year.loc[['Andhra Pradesh', 'Arunachal Pradesh',
       'Assam', 'Bihar', 'Chhattisgarh', 'Goa', 'Gujarat', 'Haryana',
       'Himachal Pradesh', 'Jharkhand', 'Karnataka',
       'Kerala', 'Madhya Pradesh', 'Maharashtra', 'Manipur', 'Meghalaya',
       'Mizoram', 'Nagaland', 'Odisha', 'Punjab', 'Rajasthan', 'Sikkim',
       'Tamil Nadu', 'Telangana', 'Tripura', 'Uttar Pradesh', 'Uttarakhand'],:]
gdp_ut = target_year.loc[['Andaman & Nicobar Islands','Chandigarh','Delhi','Jammu & Kashmir','Puducherry'],:]
gdp_india = target_year.loc[['All_India GDP'],:]
gdp_all = target_year.iloc[:-1,:]
```

```python
# Our data is cleaned and transformed and ready to use
```

### Lets analyze the growth rate for past 3 years

```python
gdp_states[[('(% Growth over previous year)','2013-14'),('(% Growth over previous year)','2014-15'),('(% Growth over previous year)','2015-16')]].plot.bar(figsize=[20,10])
plt.show()
```

###  Lets analyze the average growth percentage for each states, ut and overall country

```python
fig, ax = plt.subplots(2,2, figsize=[20,15])
gdp_all['Mean % growth (1314-1516)'].sort_values().plot.bar(ax=ax[0][0])
gdp_ut['Mean % growth (1314-1516)'].sort_values().plot.bar(ax=ax[0][1])

gdp_st['Mean % growth (1314-1516)'].sort_values().plot.bar(ax=ax[1][0])
gdp_india['Mean % growth (1314-1516)'].sort_values().plot.bar(ax=ax[1][1])

plt.tight_layout(pad=3)
plt.show()
```

```python
# Inferences -

# 1.East Indian states like Manipur, Nagaland, Tripura and Mizoram have high average growth rate.
# 2.Goa, meghalaya and odisha have very low average growth rate.
# 3.Union territories have approx similar avg growth rate except andaman and nicobar. Note 2015-16 growth rate of A&N is not available.
# 4.Rest of the states have avg growth rate 11-13% which averages out to india total gdp growth rate.
# 5.My home state (Delhi) have avg growth rate of 12% which lies little above the country avg growth rate.
```

###  Lets analyze the gdp of each states, ut and overall country for year 2015-2016

```python
fig, ax = plt.subplots(2,2, figsize=[20,15])
gdp_all['GSDP CURRENT PRICES in Crore'].sort_values().plot.bar(ax=ax[0][0])
gdp_ut['GSDP CURRENT PRICES in Crore'].sort_values().plot.bar(ax=ax[0][1])

gdp_st['GSDP CURRENT PRICES in Crore'].sort_values().plot.bar(ax=ax[1][0])
gdp_india['GSDP CURRENT PRICES in Crore'].sort_values().plot.bar(ax=ax[1][1])

plt.tight_layout(pad=3)
plt.show()
```

```python
#Inferences -

# 1.India overall gdp is approx 140 lakh crore.
# 2.5 highest gdp states - TN, UP, Karnataka, Gujarat and AP.
# 3.5 Lowest gdp states - Sikkim, AP, Meghalaya, UK, Assam.
# 4.Among UT, delhi have highest gdp but its area is relatively high. Only anomaly is J&K as its gdp is very low even though it has high area.
# 5.Poor performing areas - Arunachal Pradesh, J&K, Chattisgarh, Bihar, Jharkhand, UK. These areas covers more area but their growth and gdp are relatively low.
# Lets move on to next dataset.
```
